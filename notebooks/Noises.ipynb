# 1. Add noises

```python
# Load an image using OpenCV
image = cv2.imread('./sample.jpg')

image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

# Add non-Gaussian noise
noisy_image = add_noise(image, noise_type='exponential', scale=100,sigma=150)
noisy_image = add_noise(noisy_image, noise_type='gaussian', scale=100,sigma=150)

plt.imshow(noisy_image,cmap='gray')
plt.show()
```

# 2. Noise Combination

<p>C(n,p) = n! / p!(n-p)!<p/>

```python
noises = [
    "uniform",
    "exponential",
    "poisson",
    "gaussian",
    "salt_pepper"
]

p = 2
all_combos = np.array(list(itertools.combinations(noises, p)))
print(all_combos)
```
